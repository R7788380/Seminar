# **如何擬合多項式函數 $f(x)=6x^2-4x+1$**

#### 此為 simple_linear_regression.ipynb 裡 Exercise 的參考解答

```python
import numpy as np
import random
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.utils import plot_model
from tqdm.notebook import tqdm

import matplotlib.pyplot as plt
```

```python
# 準備資料
x_in = np.linspace(-1, 1, 200, dtype=np.float32)
y_true = 5 * x_in + np.random.rand(len(x_in))

# 資料視覺化
plt.figure(figsize=(10, 8))
plt.scatter(x_in, y_true)
plt.axis(xmin=-1, xmax=1)
plt.show()
```

```python
data = tf.data.Dataset.from_tensor_slices((x_in, y_true)).shuffle(buffer_size=200).batch(20)
```

```python
class simple_ReZero(tf.keras.Model):
    def __init__(self, alpha_init=0., w_init=1.):
        super().__init__()
        
        self.resweight = tf.Variable([alpha_init,], trainable=True)
        
#         initializer = tf.initializers.GlorotUniform()
#         self.w = tf.Variable(initializer(shape=(1,)), trainable=True)
        self.w = tf.Variable([w_init,], trainable=True)
        
    def call(self, x, resweight=True):
        if resweight:
            y = x + self.resweight * x * self.w
        else:
            y = x + x * self.w
        return self.resweight, self.w, y
```

```python
optimizer = tf.keras.optimizers.SGD(learning_rate=0.01)
```

```python
mse = tf.keras.losses.MeanSquaredError()
```

```python
steps = 300

w_inits = [1.5,2.5,3.5,4.5,5.5]
alpha_init = 0.

w_dict = {}
resweight_dict = {}
loss_dict = {}
gradient_norm_dict = {}

for w_init in tqdm(w_inits):
    model = simple_ReZero(alpha_init, w_init)
    optimizer = tf.keras.optimizers.SGD(learning_rate=0.001)
    mse = tf.keras.losses.MeanSquaredError()
    
    losses = list()
    gradients_norm = list()
    resweights = list()
    ws = list()
    
    for i in tqdm(range(steps)):
        for inp, tar in data:
            with tf.GradientTape(persistent=True) as tape:
                resweight, w, pred = model(inp, resweight=True)
                loss = mse(tar, pred)
            gradients = tape.gradient(loss, model.trainable_variables)
            optimizer.apply_gradients(zip(gradients, model.trainable_variables))

            resweights.append(resweight.numpy())
            ws.append(w.numpy())
            losses.append(loss.numpy())
            gradient_norm = np.linalg.norm([t.numpy() for t in gradients])
            gradients_norm.append(gradient_norm)
    resweights = np.array(resweights)
    ws = np.array(ws)
    
    w_dict[w_init] = ws
    resweight_dict[w_init] = resweights
    loss_dict[w_init] = losses
    gradient_norm_dict[w_init] = gradients_norm
```

```python
plt.figure(figsize=(10,6))
for w_init in w_inits:
    ws = w_dict[w_init].reshape(-1)
    resweights = resweight_dict[w_init].reshape(-1)
    losses = loss_dict[w_init]
    gradients_norm = gradient_norm_dict[w_init]
    
    norm = plt.Normalize(min(losses), max(losses))
    losses_norm = norm(losses)
    sc1 = plt.scatter(resweights, ws, c=losses_norm)
    plt.xlabel('alpha', fontsize=20)
    plt.ylabel('w', fontsize=20)
    plt.title('loss',fontsize=20)
    
#     norm = plt.Normalize(min(gradients_norm), max(gradients_norm))
#     gradients_norm_ = norm(gradients_norm)
#     sc2 = plt.scatter(resweights, ws, c=gradients_norm_)
#     plt.xlabel('alpha', fontsize=20)
#     plt.ylabel('w', fontsize=20)
#     plt.title('gradient_norm',fontsize=20)


plt.colorbar(sc1, fraction=.1)
# plt.savefig('Normal_gradient_norm.png')
plt.show()
```
